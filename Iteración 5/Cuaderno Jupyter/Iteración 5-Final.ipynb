```python
def log_assets(x):
    
    total_assets = x[['Total activo EUR Últ. año disp.','Total activo EUR Año - 1','Total activo EUR Año - 2']]

    log = pd.Series([0 for i in range(x.shape[0])],name='Size')

    for i in range(total_assets.shape[1]):
        log = log + total_assets.iloc[:,i]

    size = pd.DataFrame(np.log(log/3),columns=['Tamano'])
    
    return pd.concat([size,x],axis=1)
```

```python
def make_variables(x,columns,numerador,denominador,names):    
    # Deuda total =  Pasivo fijo + Pasivo circulante

    deuda_total = new_columns('Pasivo fijo EUR','Pasivo líquido EUR',x,'Deuda total EUR','+')
    # Activo fijo = Activo total + Activo 
    activo_fijo = new_columns('Total Activo EUR','Activo circulante EUR',x,'Activo fijo EUR','-')

    # Gastos de la explotacion = Ingresos de explotacion - Resultado de la explotacion
    gastos_explotacion = new_columns('Ingresos de explotación EUR','Resultado Explotación EUR',x,'Gastos de explotacion','-')

    x = pd.concat([x,deuda_total,activo_fijo,gastos_explotacion],axis=1)
    
    xf = x[transform_columns(columns)]
    for i in range(len(names)):
        xf = pd.concat([xf,new_columns(numerador[i],denominador[i],x,names[i],'/')],axis=1)
        
    return xf.fillna(0)
```
