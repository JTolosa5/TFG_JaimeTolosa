```python
import pandas as pd
import numpy as np
```

```python
activas1 = pd.read_csv('../Datasets/Activasv3.csv')
```

```python
activas1.shape
```

```python
activas = activas1.iloc[41500:92500,:]
```

```python
activas
```

```python
activas_part1 = activas1.iloc[0:41500,:]
activas_part1
```

```python
activas_part2 = activas1.iloc[92500:,:]
activas_part2
```

```python
for i in activas.columns[8:(len(activas.columns)-1)]:
    activas[i] = activas[i].astype(str)
```

```python
def ultimo_punto(num):
    ind = None
    for i in range(len(num)):
        if num[i] == '.':
            ind = i
            
    return ind 
```

```python
def quitar_num_sobrantes(num,ind):
    return num[:ind+4]
```

```python
quitar_num_sobrantes('213.123000',3)
```

```python
def add_zeros(num,ind):
    zeros = 3 - (len(num) - (ind+1))
    for i in range(zeros):
        num = num + '0'
    
    return num 
```

```python
add_zeros('121,5',3)
```

```python
columna = activas['Ingresos de explotación EUR Año - 1']
```

```python
def borrar_puntos(num):
    import re
    return re.sub('[.]', '', num)
```

```python
borrar_puntos('0,5')
```

```python
arr = activas['Código NIF'].values.reshape((len(activas['Código NIF'].values),1))
for j in activas.columns[8:(len(activas.columns)-1)]:
    lista=[]
    for i in activas[j]:
        if (i == 'n.s.') or (i == 'n.d.'):
            lista.append(np.nan)
        else:

            ind = ultimo_punto(i)
            if ind != None:
                if (ind+1)+3 > len(i):

                    lista.append(borrar_puntos(add_zeros(i,ind)))

                elif (ind+1)+3 < len(i) :

                    lista.append(borrar_puntos(quitar_num_sobrantes(i,ind)))
                else:
                    lista.append(borrar_puntos(i))

            else:
                lista.append(i)
    tam = len(lista)
    lista = np.array(lista).reshape((tam,1))
    arr = np.append(arr,lista,axis=1)
```

```python
colum = list(activas.columns[8:(len(activas.columns)-1)])
colum.insert(0,'Código NIF')
colum
```

```python
part2  = pd.DataFrame(arr, columns=colum)
```

```python
colum_par1=list(set(activas.columns)-set(activas.columns[8:(len(activas.columns)-1)]))
part1 = activas[colum_par1]
part1
```

```python
activas_prub = part1.join(part2.set_index('Código NIF'),on='Código NIF')
activas_prub = activas_prub[activas.columns]
```

```python
activas_finish = pd.concat([activas_part1,activas_prub,activas_part2],axis=0)
```

```python
activas_finish.to_csv('../Datasets/Activas_beta_v4.csv',index=False)
```
